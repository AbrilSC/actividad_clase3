```python
# Ejercicios de Manipulación de Datos Complejos

## Datos Simples, Manipulación de Listas y Diccionarios, Datos Complejos


#Datos Simples  ---------------------------------------

##Crear una lista de nombres de ciudades.

ciudades = ['Ushuaia', 'Rio Grande', 'Buenos Aires', 'La Plata']
print(ciudades)

##Crear un diccionario con información de empleados (nombre, cargo, salario).

empleados = {'Benjamin': {'cargo': 'Gerente', 'salario': 55000},
             'Mariana': {'cargo': 'Portero', 'salario': 40000},
             'Luis': {'cargo': 'Mesa de Entrada', 'salario': 45000}}
print(empleados)

##Mostrar cómo acceder a un elemento específico en una lista y en un diccionario.

# Acceder al tercer elemento de la lista de ciudades
print(ciudades[2])

# Acceder al salario de Luis en el diccionario de empleados
print(empleados['Luis']['salario'])

#Manipulación de Listas y Métodos ---------------------------------------

# Crea una lista inicial de números enteros
numeros = [1, 2, 3, 20, 5, 6, 7, 8, 9]

# Utiliza el método .append() para agregar un número al final de la lista
numeros.append(10)
print(numeros)

# Utiliza el método .extend() para agregar múltiples elementos a la lista
numeros.extend([11, 12, 13])
print(numeros)

# Utiliza el método .insert() para insertar un número en una posición específica de la lista
numeros.insert(3, 4)
print(numeros)

# Muestra cómo utilizar el método .sort() para ordenar la lista de números de forma ascendente
numeros.sort()
print(numeros)

# Manipulación de Diccionarios y Métodos---------------------------------------

# Crea un diccionario con información de productos
productos = {'jabon': {'precio': 50, 'stock': 10},
             'azucar': {'precio': 30, 'stock': 15},
             'agua': {'precio': 40, 'stock': 20}}

# Utiliza el método .keys() para obtener una lista de las llaves en el diccionario
print(list(productos.keys()))

# Utiliza el método .values() para obtener una lista de los valores en el diccionario
print(list(productos.values()))

# Utiliza el método .items() para obtener una lista de tuplas con pares clave-valor del diccionario
print(list(productos.items()))

# Muestra cómo utilizar el método .get() para obtener el precio de un producto específico del diccionario
print(productos.get('azucar')['precio'])

# Datos Complejos---------------------------------------

# Crear una lista de diccionarios que represente información de libros
libros = [{'título': 'Libro 1', 'autor': 'Autor 1', 'año': 2002},
          {'título': 'Libro 2', 'autor': 'Autor 2', 'año': 2006},
          {'título': 'Libro 3', 'autor': 'Autor 3', 'año': 2024}]
print(libros)

# Crear un diccionario que contenga listas como valores
estudiantes = {'Mariano': ['Programacion', 'Matematica', 'Logica'],
               'Lucas': ['Lengua', 'Geografía', 'Fisica'],
               'Sandra': ['Biología', 'Educación Física', 'Ingles']}
print(estudiantes)

# Modificar un valor en una lista dentro de un diccionario y viceversa
estudiantes['Mariano'][0] = 'Informática'
print(estudiantes)

productos['Jabon']['precio'] = 15
print(productos)
```
